```python
# Importing and Exporting Data
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Dataset/uk_universities.csv')
df
```

```python
# Melihat datatype dan spesifik pada dataset
df.info()
```

```python
# Mengecek apakah ada missing values
df.isnull().sum()
```

```python
# Dropping unnecessary columns atau hapus data yang tidak diperlukan
columns_to_drop = ['Motto', 'CWUR_score', 'Minimum_IELTS_score',
                   'UG_average_fees_(in_pounds)', 'PG_average_fees_(in_pounds)',
                   'International_students', 'Student_satisfaction',
                   'Student_enrollment', 'Academic_staff', 'Academic_Calender',
                   'Campus_setting', 'Latitude', 'Longitude', 'Website']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
# Renaming columns
df.rename(columns={
    'University_name': 'University Name',
    'Founded_year': 'Founded Year',
    'UK_rank': 'UK Rank',
    'World_rank': 'World Rank',
    'Control_type' : 'Control Type',
    'Estimated_cost_of_living_per_year_(in_pounds)': 'Estimated Cost in Pounds',
}, inplace=True)


# Tampilkan DataFrame setelah perubahan nama kolom
df
```

```python
# Ringkasan statistik
df.describe()
```

```python
# Menghitung nilai minimum
min_cost = df['Estimated Cost in Pounds'].min()
print("Nilai minimum:", min_cost)

# Menghitung nilai maksimum
max_cost = df['Estimated Cost in Pounds'].max()
print("Nilai maksimum:", max_cost)

# Menghitung nilai rata-rata
mean_cost = df['Estimated Cost in Pounds'].mean()
print("Nilai rata-rata:", mean_cost)

# Menghitung nilai tengah
median_cost = df['Estimated Cost in Pounds'].median()
print("Nilai median:", median_cost)
```

```python
# Menghitung kuartil pertama (Q1)
q1 = df['Estimated Cost in Pounds'].quantile(0.25)
print("Q1 : ", q1)

# Menghitung kuartil kedua (Q2)
q2 = df['Estimated Cost in Pounds'].quantile(0.50)
print("Q2 : ", q2)

# Menghitung kuartil ketiga (Q3)
q3 = df['Estimated Cost in Pounds'].quantile(0.75)
print("Q3 : ", q3)

# Menghitung interkuartil (IQR)
iqr = q3 - q1
print("IQR : ", iqr)
```

```python
import matplotlib.pyplot as plt

# Membuat boxplot untuk kolom UK Rank dan World Rank
df.boxplot(column=['UK Rank', 'World Rank'])
plt.title('Boxplot of UK Rank and World Rank with Outliers')
plt.show()
```

```python
# Hitung kuartil pertama (Q1)
q1 = df['World Rank'].quantile(0.25)
print("Q1 : ", q1)
# Hitung kuartil ketiga (Q3)
q3 = df['World Rank'].quantile(0.75)
print("Q3 : ", q3)
iqr = q3 - q1
# Hitung batas bawah (Lower Bound) dan batas atas (Upper Bound) untuk outlier
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Temukan outlier dalam DataFrame
outliers = df[(df['World Rank'] < lower_bound) | (df['World Rank'] > upper_bound)]

print("Outlier World Rank:")
print(outliers)
```

```python
data_world = df["World Rank"]
# Buat histogram
n, bins, patches = plt.hist(data_world, bins=5, color='yellow', edgecolor='black')

# Tambahkan Label
plt.title('Histogram Nilai')
plt.xlabel('World Rank')
plt.ylabel('Frekuensi')

# Tampilkan rentang frekuensi di sumbu x
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, ['{:.0f}-{:.0f} ' .format(bins[i], bins[i+1]) for i in range(len(bins)-1)])

# Tampilkan histogram
plt.show()
```

```python
# Mengidentifikasi baris outlier berdasarkan ciri-ciri yang unik
outliers_condition = ((df['UK Rank'] > upper_bound) | (df['UK Rank'] < lower_bound)) | (df['World Rank'] < 60) | (df['World Rank'] > 5000)

# Drop baris yang memenuhi kondisi di atas (outliers)
df_cleaned = df[~outliers_condition]  # Tanda ~ untuk negasi (tidak termasuk kondisi outlier)

# Menampilkan data setelah penghapusan outliers
print(df_cleaned.tail())
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Data
data = np.array([[76, 4608],
                 [77, 734],
                 [78, 691]])

# Inisialisasi objek MinMaxScaler
scaler = MinMaxScaler()

# Fit and transformasi data menggunakan Min-Max Scaling
scaled_data = scaler.fit_transform(data)

print("Data asli:")
print(data)

print("\nData setelah Min-Max Scaling:")
print(scaled_data)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Data contoh
data = np.array([[76, 4608],
                 [77, 734],
                 [78, 691]])

# Inisialisasi objek StandardScaler
scaler = StandardScaler()

# Fit dan transformasi data menggunakan Z-Score Transformation
z_score_data = scaler.fit_transform(data)

print("Data asli:")
print(data)

print("\nData setelah Z-Score Transformation:")
print(z_score_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data
data = pd.read_csv('/content/drive/MyDrive/Dataset/uk_universities.csv')

#Top 10 Universities in UK
plt.figure(figsize=(12, 6))
top_10 = data.nsmallest(10, 'UK_rank')
plt.barh(top_10['University_name'], top_10['UK_rank'])
plt.title('Top 10 Universities in UK')
plt.xlabel('UK Rank')
plt.ylabel('University Name')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Public vs Private Universities Distribution
plt.figure(figsize=(8, 6))
control_type_counts = data['Control_type'].value_counts()
plt.pie(control_type_counts.values, labels=control_type_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Public vs Private Universities')
plt.show()
```

```python
# Founded Year Timeline
plt.figure(figsize=(15, 6))
plt.hist(data['Founded_year'], bins=30, edgecolor='black')
plt.title('Distribution of University Founding Years')
plt.xlabel('Year Founded')
plt.ylabel('Number of Universities')
plt.tight_layout()
plt.show()
```
