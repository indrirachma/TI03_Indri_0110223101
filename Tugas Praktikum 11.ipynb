OLS

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

x = df[["Luas Bangunan", "Luas Tanah"]].to_numpy()
y = df["Harga"].to_numpy()

x = sm.add_constant(x)
model_prediksi = sm.OLS(y, x).fit();
print(model_prediksi.summary())
```

```python
x_tes = df1[["Luas Bangunan", "Luas Tanah"]].to_numpy()
y_real = df1["Harga"].to_numpy()
y_nama = df1["Nama Rumah"].to_numpy()
x_tes = sm.add_constant(x_tes)
y_pred = model_prediksi.predict(x_tes)
```

```python
df = pd.DataFrame()
df["Nama Rumah"] = y_nama
df["Harga Real"] = y_real
df["Harga Prediksi"] = y_pred
df["ERROR"]=df["Harga Real"]-df["Harga Prediksi"]
df["PERSEN PREDIKSI"]=1.0-(abs(df["ERROR"])/df["Harga Real"])
df["PERSEN"]=df["PERSEN PREDIKSI"]*100
display(df)
```

```python
df.to_csv('DatasetHasil.csv')
datasets = pd.read_csv("/content/DatasetHasil.csv")
datasets
```

```python
plt.figure(figsize=(12,6))
x = datasets.iloc[:, 0].values
plt.plot(x, datasets["Harga Real"], label = "Origin", color="b")
plt.plot(x, datasets["Harga Prediksi"], label= "Estimated", color="r")

plt.legend()
plt.xlabel('Sample ID')
plt.ylabel('Harga')
plt.title('Prediksi Harga Rumah - OLS')
plt.show()
```
