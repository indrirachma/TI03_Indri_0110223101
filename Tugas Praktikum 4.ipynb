```python
# Import library
import pandas as pd

# Membaca dataset
df = pd.read_csv('/content/drive/MyDrive/Dataset/Groceries_dataset.csv')
df.head()
```

```python
# Membuat pivot table
data = pd.crosstab(df['Member_number'], df['itemDescription'])
data
```

```python
# Ubah nilai menjadi boolean (True jika item dibeli, False jika tidak)
data = data.map(lambda x: True if x > 0 else False)
data
```

```python
from mlxtend.frequent_patterns import apriori  # Import the apriori function

# Jalankan algoritma Apriori
frequent_itemsets = apriori(data, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules # Import the association rules function

# Buat aturan asosiasi
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
# Urutkan aturan berdasarkan lift
rules = rules.sort_values('lift', ascending=False)

# Tampilkan 10 aturan teratas
rules.head(10)
```
